```python
for n in range(100,999+1):
    z = n
    z = str(z)
    z1 = int(z[0])+int(z[1])
    z2 = int(z[1])+int(z[2])
    if z1>z2:
        z = str(z2) + str(z1)
    else:
        z = str(z1) + str(z2)
    if int(z)==1115:
        print(n)
        break
```

```python
def f(x, a):
    n=''
    while x!=0:
        n = n + str(x%a)
        x//=a # x = x//2
    n = n[::-1]
    return n 
for n in range(1000,1,-1):
    z = n
    b = f(n,3)
    if z%3==0:
        b = b + b[-2:]
    if z%3!=0:
        z = z%3
        z = z*3
        b = b + f(z,3)
    if int(b,3)<=150:
        print(n)
        break
```
